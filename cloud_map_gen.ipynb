```python
import json
```

```python
json_file = open('_all_docs.json')
```

```python
data = json.load(json_file)
```

```python
author_list = []

for i in range(len(data['rows'])):
    authors = data['rows'][i]['doc']['authors']
    for author in authors:
        author_list.append(author['name'])
```

```python
print(author_list)
```

```python
author_dict = {}

for author in author_list:
    if author not in author_dict.keys():
        author_dict[author] = 1
    else:
        author_dict[author] += 1
```

```python
lis = sorted(author_dict.items(), key=lambda x: x[1], reverse=True)
sortdict = dict(lis)
print(sortdict)
```
